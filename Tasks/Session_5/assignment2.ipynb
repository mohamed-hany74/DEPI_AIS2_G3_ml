```python
import os
import random

def files(path, num):
    if not os.path.exists(path):
        os.mkdir(path)

    # create files
    for i in range(num):
        file_path = os.path.join(path, f"file_{i}.txt")
        with open(file_path, "a") as f:
            f.write(f"this is file number {i}\n")

    files_list = os.listdir(path)
    num_files = len(files_list)
    print(f"num of file {num_files}")

    numfordel = num_files // 2
    files_to_del = random.sample(files_list, numfordel)

    for f in files_to_del:
        os.remove(os.path.join(path, f))

    print(f"deleted {numfordel} files")
```

```python
files("tssass",10)
```
