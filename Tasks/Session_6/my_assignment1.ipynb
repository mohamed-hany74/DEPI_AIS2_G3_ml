# Write a Python program to remove an item from a given set

## Example
my_set: {1, 2, 3, 4, 5}

Output: {1, 2, 4, 5}

```python
class RemSet:
    def __init__(self, my_set: set, item):
        self.item = item
        self.my_set = my_set

    def removeset(self):
        my_list = list(self.my_set)

        for i in my_list:
            if self.item == i:
                my_list.remove(self.item)
                   

        self.my_set = set(my_list)
        print(self.my_set)
```

# find the factorial of a number 
## Example
(factorial of 5) = 5 * 4 * 3 * 2 * 1 = 120

```python
class Factorial :
    def __init__(self , num ):
        self.num = num

    def fnn_fact (self):
        fact = 1 
        for i in range(1,self.num+1):
            fact *= i
        return fact
```

# create a function that calculates the median of a list of numbers (a median is the middle element in sorted list of numbers)

## Example
numbers: [5, 2, 7, 3, 1]

Output: 3

```python
# Write your answer here
class Median:
    def __init__(self, numbers: list):
        self.numbers = numbers

    def get_median(self):
        nums = sorted(self.numbers)
        n = len(nums)
        mid = n // 2

        if n % 2 != 0:     
            return nums[mid]
        else:                  
            return (nums[mid - 1] + nums[mid]) / 2
```

# create a function that calculates the mode of a list of number (mode is the most frequent element in a list)

## Example
lst: [5, 2, 7, 3, 2, 1, 5]

Output: 5

```python
class Mode:
    def __init__(self, list1: list):
        self.list1 = list1
        self.max_count = 0
        self.mode_item = None

    def mode(self):
        for item in self.list1:
            count = self.list1.count(item)
            if count > self.max_count:
                self.max_count = count
                self.mode_item = item

        return self.mode_item
```

```python
m = Mode([1, 2, 2, 3, 4, 2, 5])
print(m.mode())
```

```python
def mode(lst):
    max_count = 0
    mode_item = None

    for item in lst:
        count = lst.count(item)
        if count > max_count:
            max_count = count
            mode_item = item

    return mode_item
```

# Task: take the mark from the user and convert it to grade according to the following guidelines
1. 85-100 A
2. 75-85 B
3. 65-75 C
4. 50-65 D
5. <50 F

```python
class Grade:
    def __init__(self, mark: int):
        self.mark = mark

    def get_grade(self):
        if 85 <= self.mark <= 100:
            return "A"
        elif 75 <= self.mark < 85:
            return "B"
        elif 65 <= self.mark < 75:
            return "C"
        elif 50 <= self.mark < 65:
            return "D"
        elif 0 <= self.mark < 50:
            return "F"
        else:
            return "Invalid mark"
```

```python
g = Grade(90)
print(g.get_grade())
```
