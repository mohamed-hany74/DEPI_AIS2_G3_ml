```python
import numpy as np
def smart_array_generator(mode , shape : tuple , low=-1 , high=-1) :
    if mode == 'zeros' :
        array = np.zeros(shape)
        return array
    elif mode == 'ones' :
        array = np.ones(shape)
        return array
    elif mode == 'identity' :
        array = np.identity(shape)
        return array

    elif mode == 'random' :
        array = np.random.randint(low , high , size=shape)
        return array
```

```python
print( smart_array_generator('random', (5,5),0,100))
```

```python
def apply_threshold(arr , threshold , replacment_value= -1):
    array1 = np.array(arr)
    array2 = np.where([array1 > threshold],replacment_value,array1)
    return array2
```

```python
r = [25,30,85,22,90]
apply_threshold(r,80,-1)
```
