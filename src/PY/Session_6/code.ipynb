```python
def fact (num) :
  
   if num==0 :
      return 1
    
   return    num *fact(num-1)    

```

```python
def is_prime(num) :
    if num < 2:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True
```

```python
is_prime(6)
```

```python
import random

chatbot_data = {
    "greeting": {
        "patterns": [
            "hello",
            "hi",
            "hey",
            "good morning",
            "good evening"
        ],
        "responses": [
            "Hello! How can I help you?",
            "Hi there ðŸ‘‹",
            "Hey! What can I do for you?"
        ]
    },

    "goodbye": {
        "patterns": [
            "bye",
            "goodbye",
            "see you",
            "exit"
        ],
        "responses": [
            "Goodbye!",
            "See you later!",
            "Have a nice day ðŸ˜Š"
        ]
    },

    "thanks": {
        "patterns": [
            "thanks",
            "thank you",
            "appreciate it"
        ],
        "responses": [
            "You're welcome!",
            "No problem ðŸ˜„",
            "Anytime!"
        ]
    },

    "about_bot": {
        "patterns": [
            "who are you",
            "what is your name",
            "what can you do"
        ],
        "responses": [
            "I am a simple chatbot ðŸ¤–",
            "I'm here to help you!",
            "I can answer basic questions."
        ]
    }
}


def get_response(user_input):
    user_input = user_input.lower()
    
    for intent in chatbot_data:
        patterns = chatbot_data[intent]["patterns"]
        responses = chatbot_data[intent]["responses"]
        
        for pattern in patterns:
            if pattern in user_input:
                return random.choice(responses)
    
    return random.choice(chatbot_data["about_bot"]["responses"])






```

```python
def chatbot():
    print("Chatbot: Hi! How can I help you?")
    while True:
        user_input = input("User: ").lower()
        
        if user_input in ["exit", "quit", "bye", "goodbye"]:
            print("Chatbot: Goodbye! Have a nice day! ðŸ‘‹")
            break
        
        response = get_response(user_input)
        print(f"Chatbot: {response}")
```
