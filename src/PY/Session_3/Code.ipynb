```python
def calc():
    """
    Simple calculator function.

    Allows the user to choose an operation:
    1 - Addition
    2 - Multiplication
    3 - Division
    """

    def add(num1: int, num2: int):
        return num1 + num2

    def mult(num1: int, num2: int):
        return num1 * num2

    def div(num1: int, num2: int):
        if num2 == 0:
            return "Error: division by zero"
        return num1 / num2

    print("------- calc -----------")
    print("Choose operation:")
    print("1 - Sum")
    print("2 - Multiply")
    print("3 - Divide")

    choice = input("Enter operation number: ")
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    if choice == "1":
        result = add(num1, num2)
    elif choice == "2":
        result = mult(num1, num2)
    elif choice == "3":
        result = div(num1, num2)
    else:
        result = "Invalid choice"

    print("Result:", result)


if __name__ == "__main__":
    calc()
```

```python
def password(length):
    """
    get length and creat a password
    
    """
    char = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#$%^&*"
    password = ""

    for _ in range(length):
        password += random.choice(char)

    return password
```

```python
def num_word(words):
    unique_words = set(words)
    count = {word: 0 for word in unique_words}

    for i in words:
        count[i] += 1

    return count
```

```python
def add(num1, num2):
    return num1 + num2

def sub(num1, num2):
    return num1 - num2

def div(num1, num2):
    return num1 / num2

def mult(num1, num2):
    return num1 * num2


def main():
    print("---------- calc -------------------")
    print("choose op")
    print("1 - sum")
    print("2 - sub")
    print("3 - div")
    print("4 - mult")

    try:
        choice = int(input("enter your op: "))
        num1 = float(input("enter first num: "))
        num2 = float(input("enter second num: "))

        if choice == 1:
            result = add(num1, num2)
        elif choice == 2:
            result = sub(num1, num2)
        elif choice == 3:
            result = div(num1, num2)
        elif choice == 4:
            result = mult(num1, num2)
        else:
            print("invalid choice")
            return

        print("result:", result)

    except ValueError:
        print("enter numbers only")
    except ZeroDivisionError:
        print("cannot divide by zero")
    except Exception as e:
        print(e)


if __name__ == "__main__" :
    main()
```
